```python
#Importing all the liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("2015.csv")
print(df.head())
```

```python
df.info()
```

```python
df[['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']].describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Happiness Score'], kde=True)
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Number of Countries')
plt.show()
```

Based on the histogram of the happiness score, Most of scores are around centered with the peak between 5.5 and 6.5.There are fewer countries withh very low scores.

```python
df.sort_values('Happiness Score', ascending=False)
```

```python
# Top 10 happiest country in the world with the happiness score
print(df.sort_values('Happiness Score', ascending=False)[['Country', 'Happiness Score']].head(10))
```

```python
#Top 10 bottom countries with the lowest happiness score
print(df.sort_values('Happiness Score', ascending=False)[['Country', 'Happiness Score']].tail(10))
```

```python
# Get the top 10 happiest countries
top_10_happiest = df.sort_values('Happiness Score', ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(12, 6)) # Adjust figure size for better readability
sns.barplot(x='Country', y='Happiness Score', data=top_10_happiest)
plt.title('Top 10 Happiest Countries in 2015')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot of Happiness Score vs. Economy (GDP per Capita)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Economy (GDP per Capita)', y='Happiness Score', data=df)
plt.title('Happiness Score vs. Economy (GDP per Capita)')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.show()
```

In conclusion, you should see a general upward trend, indicating that countries with higher GDP per capita tend to have higher happiness scores. This is not same always as can see some other factors affection happiness aswell.

# Summary of Findings

Based on the analysis of the 2015 World Happiness Report data:

*   The happiness scores for countries in 2015 generally range from around 2.8 to 7.6.
*   The distribution of happiness scores is somewhat bell-shaped, with **most countries having scores centered between approximately 5.5 and 6.5**. There are fewer countries with very low scores and a tapering off towards the highest scores.
*   Analysis of the top and bottom 10 countries by happiness score reveals a significant difference between the happiest nations (like Switzerland, Iceland, and Denmark) and the least happy nations (like Togo, Burundi, and Syria).
*   Examining the relationship between 'Happiness Score' and 'Economy (GDP per Capita)' through a scatter plot shows a **general upward trend**. This suggests that countries with higher GDP per capita tend to report higher happiness scores. However, the scatter in the plot also indicates that GDP per capita is not the sole determinant of happiness, and other factors included in the dataset (Family, Health, Freedom, Trust, Generosity) likely play important roles.\\
