```python
#Setup and data loading
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("supermarket_sales .csv")
print(df.head())
```

```python
#Data inspection
print(df.shape)
print(df.info())
print(df.describe())

```

```python
#Total sales for each branch

print(df.groupby('City')['Invoice ID'].count())
total_sales_per_branch = df.groupby('Branch')['Total'].sum()

print(total_sales_per_branch)
```

```python
#the branch with the greatest total sales sum

branch_with_highest_sales = total_sales_per_branch.idxmax()
print(f"The branch with the highest sales is: {branch_with_highest_sales}")
```

```python
# A bar chart showing total sales for each branch.

plt.figure(figsize=(6, 4)) # Set the size of the plot
sns.barplot(x=total_sales_per_branch.index, y=total_sales_per_branch.values)

# Label the axes and add a title
plt.xlabel("Branch")
plt.ylabel("Total Sales")
plt.title("Total Sales by Branch")

# Display the plot
plt.show()

#Best performing as per the branch is {C}
```

```python
#Calculate the average purchase total for Member customers vs Normal customers.


print(df.columns)

average_sales_by_customer_type = df.groupby('Customer type')['Total'].mean()
print(average_sales_by_customer_type)

plt.figure(figsize= (6,4)) #size of the plot
sns.barplot(x=average_sales_by_customer_type.index, y=average_sales_by_customer_type.values)
#Label of the axes and the Title name
plt.xlabel("Customer Type")
plt.ylabel("Average Sales")
plt.title("Average Sales by Customer Type")
plt.show()

# As per the barplot member spend more in the supermarket as compare with the non member.
```

```python
#Total sales according to product line
total_sales_by_product_line = df.groupby('Product line')['Total'].sum()

print(total_sales_by_product_line)
#Total highest sales as per the product line
product_line_with_highest_sales = total_sales_by_product_line.idxmax()
print(f"The product line with the highest sales is: {product_line_with_highest_sales}")
```

```python
# Create a bar chart of total sales by product line.

plt.figure(figsize=(8, 6)) # Big figure size for better readability

# Create a horizontal bar plot
sns.barplot(x=total_sales_by_product_line.values, y=total_sales_by_product_line.index)

# Label the axes and add a title
plt.xlabel("Total Sales")
plt.ylabel("Product Line")
plt.title("Total Sales by Product Line")

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()

#Food and beverages is Top performing product line
```

```python
# Use a histogram to plot the distribution of ratings.

plt.figure(figsize=(8, 5)) # Set the size of the plot

# Create a histogram using seaborn
sns.histplot(data=df, x='Rating', bins=10, kde=True) # Use 10 bins as the rating is from 1 to 10

# Label the axes and add a title
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Ratings")

# Display the plot
plt.show()
```

Based on the plot, the ratings appear to be relatively uniformly distributed across the scale from 1 to 10. There isn't a strong peak in one particular range, suggesting that customers' ratings vary widely rather than being consistently high or low.

 Observation: Customers' ratings for the supermarket tend to be spread out across the entire range from 1 to 10, indicating a varied distribution of feedback rather than a strong tendency towards high or low ratings.

Based on the analysis of the supermarket sales data, Branch C shows the highest total sales, Showing strong performance in the location. According to our analysis while comparing the customer type the member tend to spend more in the supermarket as compare to non members, It shows the membership program is working good. And in the product line food and beverages emerged as the best selling product category, contributing the most to sales.Finally on customer rating for supermarket displayed the varied distribution across the scale 1 to 10 and there is no high or low rating.
